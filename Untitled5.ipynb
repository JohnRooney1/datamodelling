```python
# Importing necessary libraries
import pandas as pd
from pycaret.regression import *

# Load the dataset from CSV
nhl_data = pd.read_csv("NHLDraftPicks2018-2023Final.csv")

# Display the first few rows of the dataset
print(nhl_data.head())
```

```python
# Step 1: Data Preparation
# (Assuming data is already clean and preprocessed)

# Step 2: Model Setup
reg_setup = setup(data=nhl_data, target='Overall', session_id=42)

# Step 3: Model Training
best_model = compare_models()

# Step 4: Model Evaluation
evaluate_model(best_model)
```

```python
from sklearn.metrics import accuracy_score
import numpy as np

# Get the predictions from the best model
predictions = predict_model(best_model)

# Calculate correlation score
correlation_score = np.corrcoef(predictions['Overall'], predictions['Label'])[0, 1]

# Convert regression problem to classification (if needed) and calculate accuracy score
predictions['Label_Rounded'] = predictions['Label'].round()
accuracy = accuracy_score(predictions['Overall'], predictions['Label_Rounded'])

print("Correlation Score:", correlation_score)
print("Accuracy Score:", accuracy)
```
